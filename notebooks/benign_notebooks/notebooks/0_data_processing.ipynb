```python
import pandas as pd

file_path = "../../data/Benign_Final/BenignTraffic.pcap.csv"

df = pd.read_csv(file_path)

df.head()
```

```python
# Ensure your DataFrame is named 'df' and contains the benign data
df.info()
```

```python
# This will show the stats for all numerical columns
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.histplot(df['Rate'], bins=50, kde=True) # kde adds a smooth line

plt.title('Distribution of Packet Rate for Benign Traffic')
plt.xlabel('Packet Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

# Let's see how many rows we have before cleaning
print(f"Original shape of the DataFrame: {df.shape}")

# Replace infinite values (both positive and negative) with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop any row that has at least one NaN value
df.dropna(inplace=True)

# Let's see the new shape after cleaning
print(f"Shape of the DataFrame after cleaning: {df.shape}")
print("\nInfinite values and rows with missing data have been removed.")
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.histplot(df['Rate'], bins=50, kde=True)
plt.title('Distribution of Packet Rate for CLEANED Benign Traffic')
plt.xlabel('Packet Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the 99th percentile for the 'Rate' column
p99 = df['Rate'].quantile(0.99)

print(f"99% of the benign traffic has a 'Rate' less than: {p99:.2f}")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df[df['Rate'] < p99]['Rate'], bins=50, kde=True)

plt.title('Distribution of Packet Rate for Benign Traffic (99th Percentile)')
plt.xlabel('Packet Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# A new histogram using a logarithmic scale for the x-axis
plt.figure(figsize=(12, 6))
sns.histplot(df['Rate'], bins=100, log_scale=True) # The magic is log_scale=True

plt.title('Distribution of Packet Rate for Benign Traffic (Log Scale)')
plt.xlabel('Packet Rate (Log Scale)')
plt.ylabel('Frequency')
plt.show()
```
